```python
import re
from collections import defaultdict
import numpy as np
import pandas
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
steps = defaultdict(dict)
```

```python
def fill(directory, x):
    with open('./%s/step-%s.xml'%(directory, x)) as f:
        for line in f:
            if 'team money' in line:
                parts = list(filter(None, re.split('[-."]', line)))
                steps[x][parts[3]] = int(parts[1])
```

```python
def plotDir(directory,totalsteps=500):
    for x in range(totalsteps):
        fill(directory,x)
    steps_arr = np.zeros((totalsteps,2))
    for x in range(totalsteps):
        steps_arr[x,0]=steps[x]['A']
        steps_arr[x,1]=steps[x]['B']
    plt.plot(range(totalsteps), steps_arr[:,0], 'b')
    plt.plot(range(totalsteps), steps_arr[:,1], 'r')
```

```python
plotDir('2018-07-07-12-52-44-2017-SampleSimulation/')
```
